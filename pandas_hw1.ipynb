Задание 5**
Вычисление метрик для мониторинга.
суммарная выручка
кол-во покупок
средний чек
среднее кол-во товаров в покупке
Дополнительно добавим возможность фильтровать данные по различным параметрам. Это может быть полезно чтобы посмотреть как меняются продажи в двух группах, или в какой-то отдельной категории товаров.
На вход функции будет подаваться датафрейм с данными о продажах, словарь с фильтрами и период, за который нужно посчитать метрики.
Функция должна вернуть датафрейм, в индексах которого будут все даты из указанного периода отсортированные по возрастанию, а столбцы - метрики ['revenue', 'number_purchases', 'average_check', 'average_number_items']. Формат данных столбцов - float, формат данных индекса - datetime64[ns].
Если в какие-то дни не было продаж, то нужно заполнить пропуск нулём.
import numpy as np
import pandas as pd
def calculate_sales_metrics(df, cost_name, date_name, sale_id_name, period, filters=None):
   """Вычисляет метрики по продажам.
   df - pd.DataFrame, датафрейм с данными. Пример
       pd.DataFrame(
           [[820, '2021-04-03', 1, 213]],
           columns=['cost', 'date', 'sale_id', 'shop_id']
       )
   cost_name - str, название столбца с стоимостью товара
   date_name - str, название столбца с датой покупки
   sale_id_name - str, название столбца с идентификатором покупки (в одной покупке может быть несколько товаров)
   period - dict, словарь с датами начала и конца периода пилота.
       Пример, {'begin': '2020-01-01', 'end': '2020-01-08'}.
       Дата начала периода входит в полуинтервал, а дата окончания нет,
       то есть '2020-01-01' <= date < '2020-01-08'.
   filters - dict, словарь с фильтрами. Ключ - название поля, по которому фильтруем, значение - список значений,
       которые нужно оставить. Например, {'user_id': [111, 123, 943]}.
       Если None, то фильтровать не нужно.
   return - pd.DataFrame, в индексах все даты из указанного периода отсортированные по возрастанию,
       столбцы - метрики ['revenue', 'number_purchases', 'average_check', 'average_number_items'].
       Формат данных столбцов - float, формат данных индекса - datetime64[ns].


```python
input_df = pd.DataFrame(
           [[820, '2021-04-03', 1, 213],
           [900, '2021-04-01', 2, 215],
           [600, '2021-04-01', 3, 220],
           [500, '2021-04-02', 4, 215],
           [400, '2021-04-02', 5, 220],
           [200, '2021-05-02', 6, [215, 208]],
           [300, '2021-05-02', 7, 220],
           [700, '2021-04-02', 8, 210],
           [600, '2021-04-01', 9, 210],
           [500, '2021-04-02', 10, 215],
           [400, '2021-04-02', 11, [220, 208]],
           [200, '2021-05-02', 12, 215],
           [300, '2021-05-02', 13, [210, 207]]],
           columns=['cost', 'date', 'sale_id', 'shop_id']
       )

input_df
```

```python
def calculate_sales_metrics(df, cost_name, date_name, sale_id_name, period=None, filters=None):
    
    def calc_cell_lenght(cell):
        """Подсчет среднего количества покупок в одном чеке"""
        aver = 0
        for i in cell:            
            if type(i) is not list:
                aver += 1            
            else:
                aver += len(i)            
        return aver/len(cell)
        
    date_list = pd.date_range(start=period.get("begin"),end=period.get("end"))  # создаем массив заданных дат
    columns = ["revenue", "number_purchases", "average_check", "average_number_items"]
    
    new_df = pd.DataFrame(index=date_list) # создаем пустой датафрейм с нужными датами
    new_df = new_df.iloc[:len(new_df.index)-1] # берем все строки кроме последней
    
    # создаем фильтры
    query = ""
    if filters is not None:
        for el in filters.keys():            
            if query != "": # если запрос уже не пустой
                query = query + "& (" + el + " == " + str(filters.get(el)) + ")"
            else:  # если запрос еще пустой
                query = "(" + el + " == " + str(filters.get(el)) + ")"   
        out_df = df.query(query) # применяем фильтры
#         print(query)
#         print(out_df)
        # группируем отфильтрованные данные
        out_df = out_df.groupby(date_name).agg({cost_name:["sum","count","mean"], "shop_id": calc_cell_lenght}).loc[period.get("begin"):period.get("end")]
    else:    
        # группируем НЕфильтрованные данные
#         out_df = df.groupby(date_name)[cost_name].agg(["sum","count","mean"]).loc[period.get("begin"):period.get("end")]
         out_df = df.groupby(date_name).agg({cost_name:["sum","count","mean"], "shop_id": calc_cell_lenght}).loc[period.get("begin"):period.get("end")]
    
    out_df.columns = columns  # именуем колонки
    
    result_df = pd.merge(new_df, out_df, left_index=True, right_index=True, how="left")  # объединяем два датафрейма по индексам
    result_df.fillna(0, inplace=True) # пустые ячейки заполняем нулями
#     result_df[["revenue","number_purchases"]] = out_df[["revenue","number_purchases"]].astype("float")  # меняем тип на float
    return result_df
```

```python
new_out = calculate_sales_metrics(input_df, "cost", "date", "sale_id", period = {"begin":"2021-04-01", "end":"2021-05-04"})
new_out
```
